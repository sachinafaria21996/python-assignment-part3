```python
# TASK 1 - File Read and Write Basics

from pathlib import Path

# Starting from a known folder next to this file
base = Path.cwd()  # Current working directory
data_dir = base / "MyFiles"
data_dir.mkdir(exist_ok=True)

# FIX: Combine the directory and filename so it saves inside "MyFiles"
file_path = data_dir / "python_notes.txt"

# Part A — Write
notes = [
    "Topic 1: Variables store data. Python is dynamically typed.\n",
    "Topic 2: Lists are ordered and mutable.\n",
    "Topic 3: Dictionaries store key-value pairs.\n",
    "Topic 4: Loops automate repetitive tasks.\n",
    "Topic 5: Exception handling prevents crashes.\n"
]

# Write mode
with open(file_path, 'w', encoding="utf-8") as file:
    file.writelines(notes)
print("File written successfully.")

# Append mode
extra_lines = [
    "Topic 6: Functions promote code reusability.\n",
    "Topic 7: File I/O allows for persistent data storage.\n"
]
with open(file_path, 'a', encoding="utf-8") as file:
    file.writelines(extra_lines)
print("Lines appended.\n")

# Part B — Read
try:
    with open(file_path, 'r', encoding="utf-8") as file:
        lines = file.readlines()
        
    print("File Contents:")
    for i, line in enumerate(lines, 1):
        print(f"{i}. {line.strip()}")
        
    print(f"\nTotal number of lines: {len(lines)}")
    
    # Search functionality
    keyword = input("\nEnter a keyword to search in the file: ").strip().lower()
    matches = [line.strip() for line in lines if keyword in line.lower()]
    
    if matches:
        print(f"\nLines containing '{keyword}':")
        for match in matches:
            print(f"- {match}")
    else:
        print(f"\nNo matches found for '{keyword}'.")

except FileNotFoundError:
    print("Error: The file was not found.")
```

```python
# Task 2
import requests

# Base URL for the API
BASE_URL = "https://dummyjson.com/products"

print("--- Step 1: Fetch and Display Products ---\n")
# Make a GET request to fetch 20 products
response = requests.get(f"{BASE_URL}?limit=20")

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    products = data.get('products', [])
    
    # Print the formatted table header
    print(f"{'ID':<4}| {'Title':<30} | {'Category':<15} | {'Price':<8} | {'Rating'}")
    print("-" * 75)
    
    # Iterate through products and print formatted rows
    for p in products:
        # Truncating titles to 30 characters to keep the table clean
        title = p['title'][:30]
        print(f"{p['id']:<4}| {title:<30} | {p['category']:<15} | ${p['price']:<7} | {p['rating']}")
else:
    print(f"Failed to fetch products. Status Code: {response.status_code}")


print("\n--- Step 2: Filter and Sort ---\n")
# Filter products with rating >= 4.5
filtered_products = [p for p in products if p.get('rating', 0) >= 4.5]

# Sort the filtered list by price in descending order
sorted_products = sorted(filtered_products, key=lambda x: x.get('price', 0), reverse=True)

# Print the filtered and sorted list
for p in sorted_products:
    print(f"- {p['title']} (${p['price']} | Rating: {p['rating']})")


print("\n--- Step 3: Search by Category ---\n")
# Make a GET request to fetch laptops
laptop_response = requests.get(f"{BASE_URL}/category/laptops")

if laptop_response.status_code == 200:
    laptops_data = laptop_response.json()
    laptops = laptops_data.get('products', [])
    
    for laptop in laptops:
        print(f"Laptop: {laptop['title']} | Price: ${laptop['price']}")
else:
    print(f"Failed to fetch laptops. Status Code: {laptop_response.status_code}")


print("\n--- Step 4: POST Request (Simulated) ---\n")
# Define the JSON body payload
post_payload = {
  "title": "My Custom Product",
  "price": 999,
  "category": "electronics",
  "description": "A product I created via API"
}

# Send the POST request to the /add endpoint
post_response = requests.post(f"{BASE_URL}/add", json=post_payload)

# Print the full response returned by the server
print("Server Response:")
import json
print(json.dumps(post_response.json(), indent=2))
```

```python
# Task 4 — Error Logger Function

import requests
from datetime import datetime
import os


def log_error(func_name, error_type, message):
    """
    Writes a timestamped error entry to error_log.txt in append mode.
    """
    # Generate the timestamp in the required format
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    # Format the log entry
    log_entry = f"[{timestamp}] ERROR in {func_name}: {error_type} — {message}\n"
    
    # Open in append mode ('a') so logs accumulate across multiple runs
    with open("error_log.txt", "a", encoding="utf-8") as file:
        file.write(log_entry)

# ==========================================
# Simulating the Errors
# ==========================================
print("--- TASK 4: LOGGING TO FILE ---\n")
print("Intentionally triggering errors to populate the log...\n")

# Trigger 1: ConnectionError (Python-level exception)
def fetch_products_bad_connection():
    try:
        print("Attempting to connect to an unreachable URL...")
        # A timeout is included so it doesn't hang forever trying to resolve the bad URL
        requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=3)
    except requests.exceptions.ConnectionError:
        print("Caught ConnectionError. Logging to file...")
        log_error("fetch_products", "ConnectionError", "No connection could be made")
    except Exception as e:
        print(f"An unexpected exception occurred: {e}")

# Trigger 2: HTTP Error (Server-level response, NOT a Python exception)
def lookup_product_not_found():
    try:
        print("Attempting to look up a non-existent product (ID 999)...")
        response = requests.get("https://dummyjson.com/products/999", timeout=5)
        
        # As requested, this is handled manually via status code, not a try-except block
        if response.status_code != 200:
            print(f"Received HTTP {response.status_code}. Logging to file...")
            log_error("lookup_product", "HTTPError", f"{response.status_code} Not Found for product ID 999")
            
    except requests.exceptions.RequestException as e:
         print(f"A network error prevented the lookup: {e}")

# Run the triggers
fetch_products_bad_connection()
lookup_product_not_found()

# ==========================================
# Read and Print Log File
# ==========================================
print("\n--- Contents of error_log.txt ---")
if os.path.exists("error_log.txt"):
    with open("error_log.txt", "r", encoding="utf-8") as file:
        print(file.read().strip())
else:
    print("Log file was not found.")
```
